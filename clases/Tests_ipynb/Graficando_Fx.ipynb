Con la funcion `partIntervalo(intervalos,n-partes)` segmentamos el intervalo en n-partes

Agregando las funciones basicas de manera externa a la clase, es posible usarlas en `listas` y arreglos de `numpy`

```python
b = partIntervalo(a,15)
c = partIntervalo(a,5)
d = partIntervalo(a,50)
print b
```

Con la función `plotFIntervalo(nIntervalos,funcion)` se genera la grafica de la funcion y los rangos de la funcion extendida a intervalos.

```python
%matplotlib inline
```

```python
outp=plotFIntevalo(b,cos),plotFIntevalo(c,cos),plotFIntevalo(d,cos)
```

```python
outp=plotFIntevalo(b,exp),plotFIntevalo(c,exp),plotFIntevalo(d,exp)
```

```python
domains = [partIntervalo(Intervalo(0.1,5.0),i) for i in [5,15,50]]
outp=plotFIntevalo(domains[0],log),plotFIntevalo(domains[1],log),plotFIntevalo(domains[2],log)
```

```python
domains = [partIntervalo(Intervalo(-np.pi/2. + 0.2,np.pi/2. - 0.2),i) for i in [5,15,50]]
outp=plotFIntevalo(domains[0],tan),plotFIntevalo(domains[1],tan),plotFIntevalo(domains[2],tan)
```

```python
domains = [partIntervalo(Intervalo(0.0,np.pi/2. - 0.2),i) for i in [5,15,50]]
outp=plotFIntevalo(domains[0],sqrt),plotFIntevalo(domains[1],sqrt),plotFIntevalo(domains[2],sqrt)
```

Finalmente en base a las funciones declaradas podemos hacer la composición para explorar combinaciones. (Varias combinaciones no jalan!: ultiplicacion por escalares, ..)

```python
def iFuncion(x):
    return log(exp(cos(x))*sqrt(x))
```

```python
domains = [partIntervalo(Intervalo(0.5,10.),i) for i in [5,15,50]]
outp=plotFIntevalo(domains[0],iFuncion),plotFIntevalo(domains[1],iFuncion),plotFIntevalo(domains[2],iFuncion)
```
